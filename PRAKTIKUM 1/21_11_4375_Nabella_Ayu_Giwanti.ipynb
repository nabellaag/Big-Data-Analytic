```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_votes = [52, 39, 78, 13, 78]
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D', '#D53032')
explode = (0,0.1,0,0,0)
```

```python
# Perhitungan total jumlah permen dalam kantong
total_candies = sum(candy_votes)
```

```python
# Perhitungan persentase peluang memilih permen Snickers
snickers_percentage = (candy_votes[candy_names.index('Snickers')] / total_candies) * 100
```

```python
#Diagram pie

plt.axis('equal')  # Membuat diagram pie menjadi lingkaran sempurna
plt.title('Persentase Peluang Memilih Permen Snickers')

plt.pie(candy_votes, 
        labels=candy_names, 
        autopct='%1.1f%%',
        colors = colors,
        explode = explode,
        shadow = True,)
plt.show()

# Menampilkan persentase peluang memilih permen Snickers
print(f'Persentase peluang memilih permen Snickers: {snickers_percentage:.2f}%')
```
